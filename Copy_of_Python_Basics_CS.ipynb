7. : 		Use a space to insert an extra space before positive numbers (and a minus sign before negative numbers)
8. :,		Use a comma as a thousand separator
9. :_		Use a underscore as a thousand separator
10. :b		Binary format
11. :c		Converts the value into the corresponding unicode character
12. :d		Decimal format
13. :e		Scientific format, with a lower case e
14. :E		Scientific format, with an upper case E
15. :f		Fix point number format
16. :F		Fix point number format, in uppercase format (show inf and nan as INF and NAN)
17. :g		General format
18. :G		General format (using a upper case E for scientific notations)
19. :o		Octal format
20. :x		Hex format, lower case
21. :X		Hex format, upper case
22. :n		Number format
23. :%		Percentage format


```python
txt = "We have {:<8} chickens."
print(txt.format(49))           # We have 49       chickens.

txt = "The temperature is {:=8} degrees celsius."
print(txt.format(-5))           # The temperature is -      5 degrees celsius.

txt = "The binary version of {0} is {0:b}"
print(txt.format(5))           # The binary version of 5 is 101

txt = "We have {:e} chickens."
print(txt.format(5))           # We have 5.000000e+00 chickens.
```

### Functions

3. Parameters & Arguments - from a function's perspective:

    b. An argument is the value that is sent to the function when it is called.

5. Keyword Arguments [kwargs] (keyword args) - key = value pairs. The order does not matter.

```python
def my_function(fname, lname):
  print(fname + " " + lname)
my_function("Emil", "Refsnes")

#------------------------------------------
# Binary args - *args (we get tuple)
def my_function(*kids):
  print("The youngest child is " + kids[2])
  print(f"number of arguments are: {len(kids)}")

my_function("Yakov", "Itzhak", "Yosef")

#------------------------------------------
# Keyword args (kwargs)
def my_function(child1, child2, child3):
  print("The youngest child is " + child3)

my_function(child1 = "Yakov", child2 = "Itzhak", child3 = "Yosef")

#------------------------------------------
# Arbitrary Keyword args (**kwargs) - we get dictionary
def my_function(**kid):
  print("His last name is " + kid["lname"])

my_function(fname = "Yakov", lname = "Israel")

#------------------------------------------
# Default Param Value
def my_function(country = "Norway"):
  print("I am from " + country)

my_function("Sweden")
my_function("India")
my_function()
my_function("Brazil")
```

#### Recursions
1. Recursion - a defined function all itself. The process of defining something in terms of itself.
2. When written correctly - recursions can be very efficient and mathematically-elegant approach to programming.
3. Be careful- it is easy to slip and have endless recursion calls.
4. Recursive algorithms must have:
    a. Stop criteria
    b. Recursive case(s)


```python
def factorial_func(number):
    assert number > 0, "Cannot calculate factorial of 0"
    print(f"factorial_func({number})")
    if number == 1:
        print("Stop Criteria reached - number = 1")
        return 1
    else:
        value = number * factorial_func(number - 1)
        print(f"returning value={value}")
        return value

# testing
for i in range(6):
    try:
        print(f"Calculating factorial of {i}")
        print(f"Calculating {i}! = {factorial_func(i)}")
        print("---------------------------------------")
    except AssertionError as msg:
        print(f"Got Assertion error !!! - message = {msg}")
        print("---------------------------------------")
        continue
```

```python
"""
List Comprehension - creating new list based on values of an existing list.
syntax:
newlist = [expression for item in iterable if condition == True]

expression is the current item in the iteration and it is also the outcome, we can manipulate it before it ends up.
condition is optional and accepts items that evaluate to True
"""
fruits = ["apple", "banana", "cherry", "kiwi", "mango"]

# creating a new list of words with a letters
newlist = [x for x in fruits if "a" in x]
print(newlist)  # ['apple', 'banana', 'mango']

# accept only numbers that are lower than 5
newlist = [x for x in range(10) if x < 5]
print(newlist)  # [0, 1, 2, 3, 4]

# manipulate expressio to upper case
newlist = [x.upper() for x in fruits]
print(newlist)  # ['APPLE', 'BANANA', 'CHERRY', 'KIWI', 'MANGO']

# set all values to "hello"
newlist = ['hello' for x in fruits]
print(newlist)  # ['hello', 'hello', 'hello', 'hello', 'hello']

# conditions over expression - manipulate the outcome
# return "orange" instead of "banana"
newlist = [x if x != "banana" else "orange" for x in fruits]
print(newlist)  ['apple', 'orange', 'cherry', 'kiwi', 'mango']
```

```python
# Sorting lists - sort()
# sorts the list alphanumerically, ascending, by default
thislist = ["orange", "mango", "kiwi", "pineapple", "banana"]
thislist.sort()
print(thislist)  # ['banana', 'kiwi', 'mango', 'orange', 'pineapple']

# sort the list descending
thislist.sort(reverse = True)
print(thislist)  # ['pineapple', 'orange', 'mango', 'kiwi', 'banana']

# customized sort function
def myfunc(n):  # sort based on distance the num from 50
  return abs(n - 50)

thislist = [100, 50, 65, 82, 23]
thislist.sort(key = myfunc)
print(thislist)  # [50, 65, 23, 82, 100]

# case insensitive sorting
thislist.sort(key = str.lower)

# sorting reverse:
thislist.reverse()
```

#### range() - sequence
1. The range() function returns a sequence of numbers, starting from 0 by default, and increments by 1 (by default) and stops before a specified number.
2. syntax: range(start, stop, step) - start - optional, default = 0, stop - required, int - not included, step - optional, default - 1.


```python
def range_sequence(x):
  _list = []
  for i in x:
    _list.append(i)
  return _list

print(range_sequence(range(3, 6)))  # [3, 4, 5]
print(range_sequence(range(3,8,1))) # [3, 4, 5, 6, 7]
print(range_sequence(range(5)))     # [0, 1, 2, 3, 4]
```

#### Tuples () - sequence:
1.   A tuple is a collection that is **ordered** and **unchangagle** (immutable) and allows duplicates
2. Tuples typically used to store collection of heterogeneous data
3. There is not append/extend/remove in tuples
4. Tuples items are indexed, start with 0.
5. It is possible to ADD tuple to a tuple
6. When creating a tuple with only one item, add ',' --> y = ("orange",)
7. It is possible to delete the tuple completely
8. Packing - Unpacking tuples
9. Join tuples - adding two tuples
10. Multiply tuples - multiplies the content given number of times using *
11. Tuple methods: count() - returns the number of times a value occurs, index(), return the position of a given value

```python
# Packing a tuple - when we create, we assign values.
_tuple = ("apple", "banana", "cherry", "apple", "watermellon")
tuple1 = ("apple", "banana", "cherry")
tuple2 = (1, 5, 7, 9, 3)
tuple3 = (True, False, False)
tuple4 = ("abc", 34, True, 40, "male")

# creating a tuple using constructor
thistuple = tuple(("apple", "banana", "cherry")) # note the double round-brackets
print(thistuple)

# Accessing items
print(tuple4[3])  # 40
print(tuple2[-2]) # 9

# ranges of indexes:
print(tuple4[1:3]) # (34, True)
print(_tuple[:3])  # ('apple', 'banana', 'cherry')

# Check if item exist:
if "male" in tuple4:
  print("found male in tuple4")

# "changing" tuple values - convert to list - change - convert to tuple
_list_tuple = list(_tuple)
print(_list_tuple)          # ['apple', 'banana', 'cherry', 'apple', 'watermellon']
_list_tuple[1] = "kiwi"
_tuple = tuple(_list_tuple)
print(_tuple)               # ('apple', 'kiwi', 'cherry', 'apple', 'watermellon')

# Adding tuple to a tuple
print(f"tuple1: {tuple1}, tuple1 id= {id(tuple1)}")
y = ("orange", )
tuple1 += y
print(f"tuple1: {tuple1}, tuple1 id= {id(tuple1)}")

# Deleting a tuple:
del _tuple
# print(_tuple) - we will get an error

# Unpacking a tuple:
# Pakcing 1
_tuple = ("apple", "banana", "watermellon")

(red, yellow, green) = _tuple
print(f"Packing 1: red={red}, yellow={yellow}, green={green}")
# red=apple, yellow=banana, green=watermellon

# Packing 2 with * - pack to a list
fruits = ("apple", "banana", "cherry", "strawberry", "raspberry")

(green, yellow, *red) = fruits
print(f"Packing 2 - green={green}, yellow={yellow}, red={red}")
# Packing 2 - green=apple, yellow=banana, red=['cherry', 'strawberry', 'raspberry']

# Looping over tuples
tuple4 = ("abc", 34, True, 40, "male")
for x in tuple4:
  print(x)

# Looping through index
for i in range(len(tuple4)):
  print(tuple4[i])

# Looping using while loop
i = 0
while i < len(tuple4):
  print(tuple4[i])
  i = i + 1

# Join two tuples
newtuple = tuple1 + tuple4
print(newtuple)  # ('apple', 'banana', 'cherry', 'orange', 'abc', 34, True, 40, 'male')

# Multiply Tuples
mul_tuple = tuple1 * 2
print(mul_tuple) # ('apple', 'banana', 'cherry', 'orange', 'apple', 'banana', 'cherry', 'orange')
```

#### Sets {} - no duplicates
1. A collection that is un-ordered (cannot determine in which order items will appear
2. Unchangable* - cannot change set items once it is create, but can remove items and add new ones only
3. Un-indexed - cannot reference by index or key

set_methods.PNG

```python
# creating set
_set = {"apple", "banana", "cherry"}
print(f"set type is: {type(set)} set values: {_set}" )  # set type is: <class 'type'> set values: {'apple', 'banana', 'cherry'}

# duplicates are not allowed
thisset = {"apple", "banana", "cherry", "apple"}
print(thisset) # {'apple', 'banana', 'cherry'}

# set data types
set1 = {"apple", "banana", "cherry"}
set2 = {1, 5, 7, 9, 34}
set3 = {True, False, False}
set4 = {"abc", 34, True, 40, "male"}

# Set constructor - need to use double round brackets
setcont = set(("apple", "banana", "cherry")) # note the double round-brackets

# Accessing set items
for x in setcont:
  print(x)

print("banana" in setcont)   # True

# Add Items
setcont.add("watermelon")
print(setcont)

# Add sets - to add items from another set into current, use update()
setcont.update(set4)
print(setcont) # {True, 34, 40, 'apple', 'banana', 'abc', 'male', 'cherry', 'watermelon'}

# Can add any iterable object (tuples, lists, dictionaries)
_list = [1,2,3,4,5,6]

setcont.update(_list)
print(setcont)  # {True, 34, 2, 3, 4, 5, 6, 40, 'apple', 'banana', 'abc', 'male', 'cherry', 'watermelon'}

# Remove Item - use remove() or discard() - if item does not exist - ERROR !!1
setcont.remove('abc')

# can use pop() method to remove - BUT - since not ordered - we do not know what item will be removed.
x = setcont.pop()
print(x)  # True

# clear() method empties the set:
setcont.clear()
print(setcont)

# del keyword will delete a cell completely.
del setcont

# Join Sets - union() method returns a new set with all items from both sets
set2 = {1, 5, 7, 9, 34}
set4 = {"abc", 34, True, 40, "male"}

set3 = set2.union(set4)
print(set3)  # {1, 34, 5, 7, 40, 9, 'abc', 'male'}

# update() method inserts the items in set2 into set1
set1 = {"apple", "banana", "cherry"}
set2 = {1, 5, 7, 9, 34}

set1.update(set2)
print(set1)  # {1, 34, 5, 7, 'apple', 'banana', 9, 'cherry'}

```
